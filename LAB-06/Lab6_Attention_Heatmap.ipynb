# Lab 6: Attention Heatmap Visualization
## Objective:
Visualize attention weights to understand how models focus on important words.

```python
# Import Libraries
import torch
import torch.nn.functional as F
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sample Sentence
sentence = ['I', 'love', 'deep', 'learning']
```

```python
# Assign Attention Weights
attention_weights = torch.tensor([0.1, 0.3, 0.4, 0.2])

# Apply Softmax Normalization
attention_weights = F.softmax(attention_weights, dim=0).detach().numpy()

print('Attention Weights:', attention_weights)
```

```python
# Plot Heatmap
sns.heatmap([attention_weights], annot=True, xticklabels=sentence, cmap='Blues')
plt.title('Attention Heatmap')
plt.xlabel('Words')
plt.ylabel('Importance')
plt.show()
```

## Conclusion:
- Attention mechanism helps focus on important words.
- Darker color means higher importance.
- Useful in NLP tasks like translation and chatbots.
