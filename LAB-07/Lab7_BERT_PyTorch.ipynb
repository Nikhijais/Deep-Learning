# Lab 7: BERT Fine-Tuning (PyTorch Version)
## Objective:
Fine-tune a pre-trained BERT model for text classification using PyTorch.

```python
# Install required libraries (run once)
# !pip install transformers torch
```

```python
# Import Libraries
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

```python
# Load Tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Sample Dataset
texts = [
    'I love this movie',
    'This film is amazing',
    'I hate this movie',
    'Worst movie ever'
]
labels = torch.tensor([1, 1, 0, 0])
```

```python
# Tokenization
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
```

```python
# Load BERT Model
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')
```

```python
# Forward Pass (Training Step)
outputs = model(**inputs, labels=labels)
loss = outputs.loss
logits = outputs.logits

print('Loss:', loss.item())
print('Logits:', logits)
```

## Conclusion:
- BERT provides high accuracy using bidirectional context.
- PyTorch version avoids TensorFlow compatibility issues.
- Suitable for NLP tasks like sentiment analysis.
